We have the data in the form of raw data in the txt file.
We cannot directly act on it and make our model creations
SO we need to properly arrange the data in the form of numpy arrays and then convert theme in teh Dataframe for further computations

The Process is as follows:
Raw Data in .txt file ---> Convert to Numpy Arrays ---> Convert to DataFrame ---> Basic Data Preprocessing( Checking for Null Values if any) ---> Apply different ML Classification models ----> Printing the Accuracy of each model

```python
with open ("Parkinsons disease raw data.txt") as f: # Opening the File in .txt format 
    lines= f.readlines() # This reads line by line from the .txt file and stores each line as a list element
    data=[]
    for line in lines: #Iterating through each line in the list
        if('phon' in line):  # We check that if this phrase comes in the line then we append it in a new list
            l=[]
            l.append(line.rstrip()) # rstrip() removes the end of line character from the string to give a complete original string
            data.append(l)
    j=0
    data3= lines[1:] # Creating another list by slicing the lines list and removing the first element
    for i in data3:
        if(j==len(data)): 
            break
        else:
            if('phon' in i): # As soon as we encounter the string with 'phon' we increase the counter as this indicates the next row
                j+=1
            else:
                data[j].append(float(i.rstrip())) # We conver the values in string to float for computations during ML model building
    print(data)

    f.close() # CLose the file we opened
   
```
